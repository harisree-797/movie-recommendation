```python
from google.colab import files
import pandas as pd

# Upload the CSV file
uploaded = files.upload()
```

```python
import io
file_name = list(uploaded.keys())[0]
movies_df = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

```python
#Create dropdown for user input
import ipywidgets as widgets
from IPython.display import display

movie_name_widget = widgets.Dropdown(
    options=['Any'] + sorted(movies_df['Movie Name'].unique().tolist()),
    description='Movie Name:',
)

language_widget = widgets.Dropdown(
    options=['Any'] + sorted(movies_df['Language'].unique().tolist()),
    description='Language:',
)

rating_widget = widgets.FloatSlider(
    value=5.0,
    min=0,
    max=10,
    step=0.1,
    description='IMDb Rating:',
)

runtime_widget = widgets.Text(
    value='Any',
    description='Running Time:',
)

year_widget = widgets.IntSlider(
    value=2000,
    min=1900,
    max=2024,
    step=1,
    description='Year of Release:',
)

maturity_widget = widgets.Dropdown(
    options=['Any'] + sorted(movies_df['Maturity Rating'].unique().tolist()),
    description='Maturity Rating:',
)

# Display
display(movie_name_widget, language_widget, rating_widget, runtime_widget, year_widget, maturity_widget)
```

```python
def filter_movies(df, movie_name, language, rating, runtime, year, maturity):
    filtered_df = df.copy()

    if movie_name != 'Any':
        filtered_df = filtered_df[filtered_df['Movie Name'] == movie_name]

    if language != 'Any':
        filtered_df = filtered_df[filtered_df['Language'] == language]

    if rating is not None:
        filtered_df = filtered_df[pd.to_numeric(filtered_df['IMDb Rating'], errors='coerce') >= rating]

    if runtime != 'Any':
        filtered_df = filtered_df[filtered_df['Running Time'] == runtime]

    if year is not None:
        filtered_df = filtered_df[pd.to_numeric(filtered_df['Year of Release'], errors='coerce') >= year]

    if maturity != 'Any':
        filtered_df = filtered_df[filtered_df['Maturity Rating'] == maturity]

    return filtered_df

filtered_movies = filter_movies(
    movies_df,
    movie_name_widget.value,
    language_widget.value,
    rating_widget.value,
    runtime_widget.value,
    year_widget.value,
    maturity_widget.value
)

# Display filtered movies
filtered_movies.head()
```

```python
def display_recommendations(filtered_df):
    if filtered_df.empty:
        print("No movies found with the given criteria.")
    else:
        display(filtered_df)

def on_change(change):
    filtered_movies = filter_movies(
        movies_df,
        movie_name_widget.value,
        language_widget.value,
        rating_widget.value,
        runtime_widget.value,
        year_widget.value,
        maturity_widget.value
    )
    display_recommendations(filtered_movies)

movie_name_widget.observe(on_change, names='value')
language_widget.observe(on_change, names='value')
rating_widget.observe(on_change, names='value')
runtime_widget.observe(on_change, names='value')
year_widget.observe(on_change, names='value')
maturity_widget.observe(on_change, names='value')

#  display
on_change(None)
```
